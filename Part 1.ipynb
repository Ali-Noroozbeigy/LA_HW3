```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
img = mpimg.imread("./part1.jpg")

imgR = img[:,:,0]
repetitionOfColor = np.zeros(256)
rangeOfColors = [i for i in range (256)]
for i in range (len(imgR)):
    for j in range(len(imgR[i])):
        repetitionOfColor[imgR[i,j]] += 1

fig = plt.figure()
ax = fig.add_axes([0,0,2,2])
ax.bar(rangeOfColors, repetitionOfColor)
```

```python
imgG = img[:,:,1]
repetitionOfColor = np.zeros(256)

for i in range (len(imgG)):
    for j in range(len(imgG[i])):
        repetitionOfColor[imgG[i,j]] += 1

fig = plt.figure()
ax = fig.add_axes([0,0,2,2])
ax.bar(rangeOfColors, repetitionOfColor)
```
